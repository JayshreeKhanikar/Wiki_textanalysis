## defining function to return categories and pages

```python
def category_call(Category):
    """
    function takes in Category as parameter and created a param dictionary for the request.get method to run WikiAPI.
    
    Params
    ------
    Category : str
        Name of the category to be scraped from wikiAPI.
         
    Returns
    -------
    DataFrame
        Pandas DataFrame containing the category from wiki categories.
    """
    param = {
        'action': 'query',
        'format': 'json',
        'list': 'categorymembers',
        'cmtitle': 'Category:{}'.format(Category),
        'cmlimit': 'max'
    }
    
    r = requests.get('https://en.wikipedia.org/w/api.php?', params = param) 
    return (r.json())['query']['categorymembers']
    #pd.DataFrame(r.json()['query']['categorymembers'])
```

## Defining function that takes try and except and clean the text:

```python
def category_page2(category, collection, max_depth = 0):
    """
    function takes in Category and empty list as parameters. It calls category_call() function internally with category
    parameter checks for title that have 'Category:" as part of string. It calls page_call() function interenally on 
    title that does not have "Category:". a list of title that has "Category:" is created and each member is used to 
    recurse through the catgory_page1 function untill all the pages for category are stored in the dictionary.
    
    Params
    ------
    Category : str, list
        Name of the category to be scraped from wikiAPI.
        empty list to store dictionary of pageid and title
         
    Returns
    -------
    List
        List of dictionaries containing pageid, title from wiki pages.
    """
    max_depth += 1
    if (max_depth < 6):
        print(category, max_depth)
        category1 = category_call(category)

        sub_cat =[]
        cat_list=[]

        for item in category1:
            #print(item)
            cat= {}

            try:
                if "Category:" not in item['title']:
                    cat['pageid'] = item['pageid']
                    cat['title'] = item['title']
                    cat['category'] = category
                    #print(cat['title'])
                    content = page_call(cat['title'])
                    cat['content'] = page_call(cat['title'])
                    # connect
                    cat_list.append(cat)
            except:
                if "Category:" not in item['title']:
                    cat['pageid'] = item['pageid']
                    cat['title'] = ''
                    cat['category'] = category
                    #print(cat['title'])
                    content = page_call(cat['title'])
                    cat['content'] = page_call(cat['title'])
                    # connect
                    cat_list.append(cat)

            if "Category:" in item['title']:
                item['title'] = item['title'].replace('Category:', '')
                sub_cat.append(item['title'])

        #batch insert of cat_list if list is not empty
        if len(cat_list)>0:
            collection.insert_many(cat_list)

#         if len(sub_cat) < 1:
#             return cat_list

        for sub in sub_cat:
            cat_list = category_page2(sub, collection, max_depth)
    #     return cat_list
```

## Creating function to find similar document:

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import TruncatedSVD
```

```python
def doc_matrix(df, array, min_df, search_query, n_comp ):
    tfidf = TfidfVectorizer(ngram_range=(1,2), min_df = min_df)
    array_tf = tfidf.fit_transform(array)
    array_tf_df =  pd.DataFrame(array_tf.toarray(), index=df.pageid, columns=tfidf.get_feature_names())
    #return array_tf_df.shape
    
    query = pd.Series(search_query)
    query_tf = tfidf.transform(query)
    query_tf_df = pd.DataFrame(query_tf.toarray(), 
                                       index=query, 
                                       columns=tfidf.get_feature_names())
    query_df_tf_df = array_tf_df.append(query_tf_df)
    
    svd = TruncatedSVD(n_components= n_comp)
    component_names = ["component_"+str(i+1) for i in range(n_comp)]
    svd_matrix = svd.fit_transform(query_df_tf_df)
    svd_df = pd.DataFrame(svd_matrix, index=query_df_tf_df.index, columns=component_names)
    
    search_term_svd_vector = svd_df.loc[query_tf_df.index]
    svd_df['cosine_sim'] = cosine_similarity(svd_df, search_term_svd_vector)
    return svd_df[['cosine_sim']].sort_values('cosine_sim', ascending=False).head(5)
    
```
